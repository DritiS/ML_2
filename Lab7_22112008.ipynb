```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['Attrition']),df['Attrition'],test_size=0.2,random_state=42)
```

```python
categorical = ['BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus', 'OverTime']
```

```python
from sklearn.preprocessing import FunctionTransformer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
trf1 = ColumnTransformer([
    ('ohe',OneHotEncoder(handle_unknown='ignore'),categorical)    
], remainder ='passthrough')
```

```python
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
rf = RandomForestClassifier(oob_score=True)
```

```python
from sklearn.pipeline import Pipeline
pipe = Pipeline([
    ('trf1',trf1),
    ('rf',rf)
])
```

```python
pipe.fit(X_train,y_train)
```

```python
rf.oob_score_
```

```python
y_pred = pipe.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(pipe, X_train, y_train, cv=5, scoring='accuracy')
```

```python
param_grid = {
    "rf__n_estimators": [20,60,100,120],
    "rf__max_features":[0.2,0.6,1.0],
    "rf__max_depth":[2,8,None],
    "rf__max_samples":[0.5,0.75,1.0]
}
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(pipe, param_grid, cv=5, scoring='accuracy')
grid.fit(X_train, y_train)
```
